```python
# variables categóricas vs variables numéricas

print(f"Categóricas{list(df.select_dtypes(exclude = 'number').columns)}")
print(f"numéricas {list(df.select_dtypes(include='number').columns)}")
```

```python
# limpieza de datos, utilizamos la función isna para detectar datos faltantes, independientemente 
# de si son Nan o nulos.

pd.DataFrame({'column name':df.columns,
                'sum':df.isnull().sum(),
                'percentage of nulls':df.isnull().mean()*100})
```

```python
# limpieza de datos

pd.DataFrame({'column name':df.columns,
                'sum':df.isna().sum(),
                'percentage of nulls':df.isna().mean()*100})
```
